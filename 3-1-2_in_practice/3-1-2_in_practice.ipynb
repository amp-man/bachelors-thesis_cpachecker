```python
import pandas as pd
import seaborn as sns
import matplotlib.ticker as ticker

sns.set_theme()
```

```python
itlimit_max_900s_dirty = pd.read_csv(filepath_or_buffer='./run1_5_900s_itlimit_max_safetyzone.table.csv',sep='\t',header=2,index_col=0)
itlimit_average_900s_dirty = pd.read_csv(filepath_or_buffer='./run1_3_900s_itlimit_average_safetyzone.table.csv',sep='\t',header=2,index_col=0)
noitlimit_700s_dirty = pd.read_csv(filepath_or_buffer='./run1_4_700s_noitlimit_safetyzone.table.csv',sep='\t',header=2,index_col=0)
```

```python
data_in = itlimit_max_900s_dirty.copy()

#Drop completely empty columns
itlimit_max_900s = data_in.dropna(axis='columns',how='all')
```

```python
data_in = itlimit_average_900s_dirty.copy()

#Drop completely empty columns
itlimit_average_900s = data_in.dropna(axis='columns',how='all')
```

```python
data_in = noitlimit_700s_dirty.copy()

#Drop completely empty columns
noitlimit_700s = data_in.dropna(axis='columns',how='all')
```

```python
cd_average = pd.concat([noitlimit_700s, itlimit_average_900s], keys=['run1_700s_noitlimit_safetyzone..', 'run1_900s_itlimit_average_safetyzone..'], names=['Run Name'])
```

```python
ax = sns.scatterplot(data=cd_average, x='Number of iterations', y='CPU Time (s)', hue='Run Name', style='Run Name')
ax.axhline(y=800, color=sns.color_palette()[0], linestyle='--', linewidth=0.9, alpha=0.5)
ax.axvline(x=2103956, color=sns.color_palette()[1], linestyle='--', linewidth=0.9, alpha=0.5)
ax.xaxis.set_major_formatter(ticker.EngFormatter())
sns.move_legend(ax, "lower right")

#ax.get_figure().savefig('3-1-2_runcomparison_average.pdf', bbox_inches = 'tight')
```

```python
cd_max = pd.concat([noitlimit_700s, itlimit_max_900s], keys=['run1_700s_noitlimit_safetyzone..', 'run1_900s_itlimit_max_safetyzone..'], names=['Run Name'])
```

```python
ax = sns.scatterplot(data=cd_max, x='Number of iterations', y='CPU Time (s)', hue='Run Name', style='Run Name')
ax.axhline(y=800, color=sns.color_palette()[0], linestyle='--', linewidth=0.9, alpha=0.5)
ax.axvline(x=16119332, color=sns.color_palette()[1], linestyle='--', linewidth=0.9, alpha=0.5)
ax.xaxis.set_major_formatter(ticker.EngFormatter())
sns.move_legend(ax, "lower right")

#ax.get_figure().savefig('3-1-2_runcomparison_max.pdf', bbox_inches = 'tight')
```
